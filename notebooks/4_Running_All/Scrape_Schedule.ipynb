### Scrape Game Schedule


```python
jan_url = 'https://www.basketball-reference.com/leagues/NBA_2023_games-january.html'
jan = pd.read_html(jan_url)[0]
jan
```

```python
feb_url = 'https://www.basketball-reference.com/leagues/NBA_2023_games-february.html'
feb = pd.read_html(feb_url)[0]
feb
```

```python
march_url = 'https://www.basketball-reference.com/leagues/NBA_2023_games-march.html'
march = pd.read_html(march_url)[0]
march
```

```python
april_url = 'https://www.basketball-reference.com/leagues/NBA_2023_games-april.html'
april = pd.read_html(april_url)[0]
april
```

```python
# append all together
schedule = jan.append(feb).append(march).append(april)
schedule
```

```python
# turn date into datetime
schedule['Date'] = pd.to_datetime(schedule['Date'])
```
