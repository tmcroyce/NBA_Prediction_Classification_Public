```python
url= 'https://www.nba.com/stats/teams'
driver = webdriver.Chrome()
```

```python

driver.get(url)
# # find by class StatsTeamsList_divContent__JvxSY
soup = BeautifulSoup(driver.page_source, 'html.parser')

# get all links
links = soup.find_all('a', href=True)
# keep only links and link text
links = [[link['href'], link.text] for link in links]
# keep only links with /stats/team/ in them
links = [link for link in links if '/stats/team/' in link[0]]
# save to dataframe
links = pd.DataFrame(links, columns=['link', 'team'])
links
```

```python
links['url2'] = links['link'].str[6:]
links
```

```python
links['url'] = 'https://www.nba.com' + links['url2']
# if link ends with /, remove it
links['url'] = links['url'].apply(lambda x: x[:-1] if x[-1] == '/' else x)

# remove duplicates
links = links.drop_duplicates(subset='url')
links = links.sort_values('team')
links

# find by class StatsTeamsList_divContent__JvxSY
```

```python
links['team_name'] = links['team'].apply(lambda x: x.split(' ')[-1])
links
```

```python
links['fixed_url'] = links['url']+ '/' + links['team_name'].str.lower()
```

```python
all_roster_links = links['fixed_url'].to_list()
all_roster_links
```
