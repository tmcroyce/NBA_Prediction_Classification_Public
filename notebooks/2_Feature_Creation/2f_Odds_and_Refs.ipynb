###  Set up GameID to connect with main testing df
bigdataball uses a different form of gameids than Nba.com or BBallRef.  


```python
# get opposing team abbrev
def get_opp(df,  game_id, team_abbrev):
    data = df[df['GAME-ID'] == game_id]
    opp_abbrev = data[data['Nba_com_team_Abbrev'] != team_abbrev]
    return opp_abbrev['Nba_com_team_Abbrev'].values[0]
```

```python
df['opponent'] = df.apply(lambda row: get_opp(df, row['GAME-ID'], row['Nba_com_team_Abbrev']), axis=1)
df.head()
```

```python
df['date'] = df['DATE'].astype(str).str.replace('/', '_')
df.head()
```

```python
def make_gameid(df, gameid, team_abv,  home_or_road, date):
    data = df[df['GAME-ID'] == gameid]
    home_team_data = data[data['VENUE'] == 'H']
    home_team = home_team_data['Nba_com_team_Abbrev'].values[0]
    away_team_data = data[data['VENUE'] == 'R']
    away_team = away_team_data['Nba_com_team_Abbrev'].values[0]
    home_gameid = home_team + ' vs. ' + away_team + '_' + date
    away_gameid = away_team + ' @ ' + home_team + '_' + date
    if team_abv == home_team:
        return home_gameid
    elif team_abv == away_team:
        return away_gameid
```

```python
df['Matchup_GameDate'] = df.apply(lambda row: make_gameid(df, row['GAME-ID'], row['Nba_com_team_Abbrev'], row['VENUE'], row['date']), axis=1)
df.head()
```

```python
df = df.drop(columns = ['BOX SCORE\nURL', 'ODDS\nURL', 'HALFTIME' ])
```
