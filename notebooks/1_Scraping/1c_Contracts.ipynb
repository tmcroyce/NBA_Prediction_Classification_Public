```python
# append the data
appended_contracts = []

for file in pf[0]:
    data = pd.read_excel('player_contracts_outputs\\' + str(file))
    dat = pd.DataFrame(data)
    appended_contracts.append(dat)

appended_contracts = pd.concat(appended_contracts)
appended_contracts
```
